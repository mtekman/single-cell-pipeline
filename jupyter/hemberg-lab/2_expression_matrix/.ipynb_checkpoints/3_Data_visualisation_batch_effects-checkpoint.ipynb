```r
#### Data Visualisation ####

# We now need to see what happened to the expression matrix after the quality control step we just did.
# We also need to control for batch effects - technical artefacts that occur in sample handling, such that 
# two sets of samples preperated in different labs/days will differ more than those handled together.
#   - Batch effects can be mistaken for biological  variation, so we need to look for batch effects in a controlled
#     manner
#

# Let's first load the data we saved earlier
library(SingleCellExperiment)
library(scater)
options(stringsAsFactors = FALSE)
umi <- readRDS("tung/umi.rds")
umi.qc <- umi[rowData(umi)$use, colData(umi)$use]
endog_genes <- !rowData(umi.qc)$is_feature_control
```

```r
# Let's do a PCA plot
# - convert a set of observations into a set of values of linearly uncorrelated variables called Principle Components.
#   The number of PCs is <= number of original variables
# - PCs → Eigenvectors of the covariance matrix (how ith and jth variable are linearly related (e.g. via Pearson correl…))
#        - Eignevectors are sorted by eigenvalue so that first two PCAs show the most variability, with each succeeding
#          PC being orthogonal to the previous PC

#attach(mtcars)
#par(mfcol=c(1,2))

# PCA: Before QC, Before Log-transform
plotPCA(
    umi[endog_genes, ],
    exprs_values = "counts",
    colour_by = "batch",
    size_by = "total_features",
    shape_by = "individual"
)

# PCA: Before QC, After Log-transform
plotPCA(
    umi[endog_genes, ],
    exprs_values = "logcounts_raw",
    colour_by = "batch",
    size_by = "total_features",
    shape_by = "individual"
)

# We can see that log transformation reduces the variance on the first PC (component 1 76% → 463%), and already seperates
# some of the biological effects
# - It also makes the distribution of the expression values more normal.
# - We will proceed with using log transformed counts.

# Note: Use logcounts and not logcounts_raw in SingleCellExperiment, as the former normalises the data by the library size
#       (e.g CPM "counts-per-million" normalisation), and accounts for more technical factors such as sequencing depth.
# - We will proceed with logcounts_raw only for demonstration purposes
```

```r
# PCA: After QC, log counts
plotPCA(
    umi.qc[endog_genes,],
    exprs_values = "logcounts_raw",
    colour_by = "batch",
    size_by = "total_features",
    shape_by = "individual"
)

# We now have much less variance in PC1 (17%) and we can see that the previous orange NA19098.r2 batch no longer form a 
# group of outliers.
```

```r
# By default we only use the top 500 most variable genes to calculate the PCA, but we can include our entire set of genes
# if need be:

# All genes
plotPCA(
    umi.qc[endog_genes,],
    ntop = nrow(umi.qc), # num genes
    exprs_values = "logcounts_raw",
    colour_by = "batch",
    size_by = "total_features",
    shape_by = "individual"
)

plotPCA(
    umi.qc[endog_genes,],
    ntop = 50,
    exprs_values = "logcounts_raw",
    colour_by = "batch",
    size_by = "total_features",
    shape_by = "individual"
)
```

```r
# We can use tSNE as another method to cluster our matrix, but this has the following pitfalls:
# - tSNE is stochastic, and so needs to be seeded the same way to get consistent results, leading to multiple different
#   interpretations
# - It functions by a perplexity 'k' parameter which attempts to divide the data into K cells. Ideally the K parameter
#   should be much smaller than the perplexity.
# - It needs multiple iterations to reach stability ~ 5000
# - Size of clusters actually mean nothing, since the same cluster can drastically change shape over a different amount
#   of generations.
# - Distances between clusters typically do not mean anything either.
# - The perplexity value must scale with the number of points used to preserve the overall global geometry of a plot
# - You really need more than one plot to get a good idea of what is going on.
# - It can definitely find pattens where other dimensional-reduction algorithms cant but it is too flexible and
#   is far too tricky to interpret --- AVOID
```

```r
### Identifying confounding factors ###
#
# It may be best to plot each component against detected genes and look at the R^2 correlation
plotQC(
    umi.qc[endog_genes, ],
    type = "find-pcs",
    exprs_values = "logcounts_raw",
    variable = "total_features"
)

# From this we can see that PC1 is represented by the total amount of detected genes.

```

```r
# We can also the marginal R² value for each variable when fitting a linear regression model  for the expression values
#  for each gene against just that variable, and display a density plot of gene-wise marginal R² values

plotQC(
    umi.qc[endog_genes, ],
    type = "expl",
    exprs_values = "logcounts_raw",
    variables = c(
        "total_features",
        "total_counts",
        "batch",
        "individual",
        "pct_counts_ERCC",
        "pct_counts_MT"
    )
)

# From this we can see that the detected number of genes (total_features) and the sequencing depth (total_counts) are
# very good at explaining the variance in the matrix, and so we can use these two variables as good candidates for
# conditioning out in a normalisation steo, or including in downstream stats models.
# - ERCC expression also seems to be quite important
# - curiously, batch seems to explain more than individual -- hnting that there is definitely a batch effect that
#   overrides the strength of individual cell expression.

# Other than batch correction, we can also compensate using a tool called scLVM which helps with the identification and
# subtraction of cell-cycle or apoptosis processes.

# We must account for these types of confounding factors via: Normalisation
#
```

```r
#### Normalisation ####

# Library size alludes to the types of cells(?) accomodated/selected for by the sequencing protocol. The size
# of the library must be taken into account when determining gene expression
# - but some quantification tools do this already and so don't need this type of normalisation (Cufflinks, RSEM)

# Most normalisation methods can be taken from bulk RNA-seq such as (UQ,SF,CPM,RPKM,FPKM,TPM)
# - CPM is the easiest way - You take each column (sample) and divide each element in that column by the column total * 1M.
#   But you must exclude spike ins from this calculation so that we can still correct for total RNA content. So we will
#   only use the endogenous genes.

calc_cpm <- function(expr_mat, spikes=NULL){
    norm_factor <- colSums(expr_mat[-spikes, ])
    return(t(t(expr_mat)/norm_factor)) * 10^6
}

# Problem of this approach is that if your sample contains genes that both very highly expressed and also differentially
# expressed across the cekks, then the total molecules in a cell may depend on whether that genes is on/off in the cell
# and so normalising by total molecules could hide this differential expression of those genes, or falsely create
# differential expression in the remaining genes

# - Other methods such RPKM, FPKM, and TPM are just variants on CPM which further adjust the counts by the length of the
#   respective gene/transcript.
# 
# One package that is optimises CPM for single-cell is the `scran` package which is good for large numbers of zero values
# per cell by pooling cells together and alculating a normalisation factor (similar to CPM) for the sum of each pool.
# Since each cell is found in many different pools, cell-specific factors have to be deconvoluted from the collection of
# pool specific factors using linear algebra.

# Three better methods we can use instead is:
#  SF "size factor" which is used by DESeq, a.k.a RLE "relative log expression" in edgeR
#  - You take the geometric mean of each gene across all cells. The SF for each cell is the median across genes to the
#    ratio of the expression to the gene's geometric mean.
#  - One drawback to this method is that since it uses the geometric mean, only genes with non-zero expression across all
#    cells can be used in the calculation, making NOT ideal for low-depth scRNASeq expreiments.
# 
calc_sf <- function(expr_mat, spikes=NULL){
    geomeans <- exp(rowMeans(log(expr_mat[-spikes, ])))
    # exponential of the mean log value of all cells for a given gene
    SF <- function(cnts){
        median((cnts/geomeans)[(is.finite(geomeans) & geomeans > 0)])
        # the median of a row of counts divided by the geometric mean on the condition that the geomeans is non-zero
        # for a particular gene
    }
    norm_factor <- apply(expre_mat[-spikes, ], 2, SF)
    return(t(t(expr_mat)/norm_factor))
}

# UQ "upper quartile" where you divide each column (sample) by the 75% quantile of counts for each library. This quantile
# is scaled by the median across cells to keep the absolute  level of expression reasonable consistent.
# - Again, for low-depth scRNA experiments the large number of undetected genes could results in the 75% quantile being 0
#   But we can overcome this by using a higher quantile or excluding zeros before the 75% quantile calc.

calc_uq <- function(expr_mat, spikes=NULL){
    UQ <- function(x){
        quantile(x[x > 0], 0.75)
    }
    uq <- unlist(apply(expr_mat[-spikes, ], 2, UQ))
    norm_factor <- uq /median(uq)
    return(t(t(expr_mat)/norm_factor))
}

# TMM "weighted trimmed mean of M-values" - where M-values are log2 fold changes between individual cells for a given gene.
# - You use one cell as a reference, and you calculate M-values for all other cells to this reference, removing the
#   top and bottom ~30% and then averaging the remaining values.
# - All non-reference cells are multiplied by the this factor, but this might not be good if you have not enough
#   non-zero genes left behind after trimming.

```
