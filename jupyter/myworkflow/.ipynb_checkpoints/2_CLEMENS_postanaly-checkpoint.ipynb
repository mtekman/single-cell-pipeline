```r
# this file will create a sce object for data that came from umi_tools count matrix

library(scater)
library(SingleCellExperiment)

# might have to remove the first word "genes" from the first line in the counts matrix.
# the next lines will read it in and set proper gene names
dset = 11
```

```r
counts_raw <- read.table(paste("/extra/sebastian_arnolds/analysis/trim/star_", dset, "/counts_matrix.tsv", sep=""), sep="\t")
```

```r
bc <- read.table(paste("/extra/sebastian_arnolds/analysis/celseq_barcodes_raw_", dset, ".txt", sep=""))
bc <- as.vector(bc[, 1])
```

```r
counts <- counts_raw[, bc]
cm <- as.matrix(counts)
```

```r
# i exchanged some gene names quick and dirty to get these as labels in the plots
rownames(cm)[which(rownames(cm) == "ENSMUSG00000062327")] <- "Brachyury"
rownames(cm)[which(rownames(cm) == "ENSMUSG00000032446")] <- "Eomes"
rownames(cm)[which(rownames(cm) == "ENSMUSG00000030544")] <- "Mesp1"
rownames(cm)[which(rownames(cm) == "ENSMUSG00000024406")] <- "Pou5f1"
rownames(cm)[which(rownames(cm) == "ENSMUSG00000012396")] <- "nanog"
rownames(cm)[which(rownames(cm) == "ENSMUSG00000096014")] <- "Sox1"
rownames(cm)[which(rownames(cm) == "ENSMUSG00000074637")] <- "Sox2"
rownames(cm)[which(rownames(cm) == "ENSMUSG00000090125")] <- "Pou3f1"
rownames(cm)[which(rownames(cm) == "ENSMUSG00000060206")] <- "zfp462"
rownames(cm)[which(rownames(cm) == "ENSMUSG00000031297")] <- "slc7a3"

sce <- SingleCellExperiment(assays = list(counts = cm))
# a command to get gene names in general.. now need to find out how to use them in plots
sce <- getBMFeatureAnnos(sce, 
                         filters = "ensembl_gene_id", 
                         attributes = c("ensembl_gene_id", "external_gene_name"), 
                         feature_symbol = "mgi_symbol",
                         feature_id = "ensembl_gene_id",
                         biomart = "ENSEMBL_MART_ENSEMBL",
                         dataset = "mmusculus_gene_ensembl", 
                         host = "www.ensembl.org"
                        )

# create a log2 table
assay(sce, "logcounts") <- log2(calculateCPM(sce, use.size.factors = FALSE) + 1)
```

```r
# calculate QCMetrics on a sce object
sce <- calculateQCMetrics(sce, exprs_values = "counts")

# expression plot of specific genes
plotExpression(sce, c("Eomes", "Brachyury", "Pou5f1", "nanog", "Sox2", "Pou3f1", "zfp462", "slc7a3"), x="Eomes")


# expression plot of all cells
plotQC(sce, type="highest-expression")
```
