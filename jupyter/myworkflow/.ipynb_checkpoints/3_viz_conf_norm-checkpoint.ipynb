```r
## NORMALIZATION!

# Let's define a few common methods suitable for our data:
# - PCA defaults
plotPCA_defaults <- function(data){
    plotPCA(
        data,
        colour_by = "outlier",
        size_by = "total_features",
        shape_by = "filter_highgenes"
    )
}

# - normaliseExpr defaults
normaliseVia = function(data, meth, ...) {
    return(copy(
        normaliseExprs(
            data,
            method = meth,
            ...,
            return_log = TRUE,
            return_norm_as_exprs = TRUE
        )
    ))
}
```

```r
# CPM
# - danger: highly expressed genes that are differentiially expressed across cells max
#           become hidden by this technique. 
umi.cpm <- log2(calculateCPM(umi, use.size.factors = FALSE) + 1)

# RLE(SF) (DESeq)
# - danger: geometric mean of gene across cells is used, so non-zero expr. values will
#           be used in the calculation -- unideal for low-depth RNA-seq.
umi.rle <- normaliseExprs(umi, method = "RLE", 
                          return_log = TRUE, return_norm_as_exprs = TRUE)

# UQ
# - danger: for low-depth seq, the large amount of undetected genes could result in the
#           75% quantile being zero. We can get around this by using the 99% quantile,
#           or excluding zeros prior to quant.
umi.uq99 <- normaliseExprs(umi, method = "upperquartile", p = 0.99, 
                           return_log = TRUE, return_norm_as_exprs = TRUE)
umi.uq75 <- normaliseExprs(umi, method = "upperquartile", p = 0.75, 
                           return_log = TRUE, return_norm_as_exprs = TRUE)
umi.uq50 <- normaliseExprs(umi, method = "upperquartile", p = 0.50, 
                           return_log = TRUE, return_norm_as_exprs = TRUE)

# TMM
umi.tmm <- normaliseExprs(umi, method = "TMM",
                              return_log = TRUE, return_norm_as_exprs = TRUE)

# Raw log value, just for reference
umi.raw <- copy(umi)
logcounts(umi.raw) <- log2(counts(umi) + 1)
```

```r
# SCRAN -- CPM designed for using in single-cell, pools zero-value cells together.
qclust <- quickCluster(umi, min.size = 30)
umi.scran <- computeSumFactors(umi, sizes = 15, clusters = qclust)
umi.scran <- normalize(umi.scran)
```

```r
# Downsample
downSampleMatrix <- function (expr_mat){
    min_lib_size <- min(colSums(expr_mat))
    down_sample <- function(x) {
        prob <- min_lib_size/sum(x)
        return(unlist(lapply(x, function(y) {
            rbinom(1, y, prob)
        })))
    }
    down_sampled_mat <- apply(expr_mat, 2, down_sample)
    return(down_sampled_mat)
}
umi.downsamp = copy(umi)

# Log-counts yields a lot of zeroes, let's see counts too
counts(umi.downsamp) <- downSampleMatrix(counts(umi.downsamp))
logcounts(umi.downsamp) <- log2(counts(umi.downsamp) + 1)
```

```r
tableau <- rbind(
    head(logcounts(umi.raw), 1),
    head(logcounts(umi.uq99), 1),
    head(logcounts(umi.uq75), 1),
    head(logcounts(umi.uq50), 1),
    head(logcounts(umi.rle), 1),
    head(logcounts(umi.tmm), 1),
    head(logcounts(umi.scran), 1),
    head(logcounts(umi.downsamp), 1)
)
rownames(tableau) <- c("RAW", "UQ99", "UQ75", "UQ50", "RLE", "TMM", "SCRAN", "Down.")

message("Log comparison of the different normalisation methods across the same gene.")
tableau
```
