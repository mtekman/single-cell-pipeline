```r
library(scRNA.seq.funcs)
library(matrixStats)
library(M3Drop)
library(RColorBrewer)
library(scran)
library(SingleCellExperiment)
set.seed(1)
```

```r
umi <- readRDS("post_common_normalisation_approaches.rds")
umi <- computeSumFactors(umi)
umi <- normalize(umi)
```

```r
Brennecke_HVG <- BrenneckeGetVariableGenes(
    counts(umi),         # do NOT use logcounts
    fdr = 0.01,          # changing to 0.1, 0.5 yields no changes
    minBiolDisp = 0.5    # changing to 0.1, 0.01 yields no changes
)

HVG_genes <- Brennecke_HVG$Gene

# store the HGV
rowData(umi)$bren_HGV <- rownames(umi) %in% HVG_genes

message("Here we can see (pink) genes with significant biological variability, after multiple-testing correction")
```

```r
# We can also use M3Drop
M3Drop_genes <- M3DropFeatureSelection(
    counts(umi),
    mt_method = "fdr",
    mt_threshold = 0.01
)

rowData(umi)$m3_HGV <- rownames(umi) %in% M3Drop_genes$Gene
```

```r
M3DropExpressionHeatmap(
    M3Drop_genes,
    counts(umi)
    #cell_labels = 
)

Jacc <- sum(M3Drop_genes$Gene %in% HVG_genes)/length(unique(c(M3Drop_genes$Gene, HVG_genes)))
message("Jaccard similarity index between M3Drop and HVG_genes is ", Jacc, ", which is low.")
```
