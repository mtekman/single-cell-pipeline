```r
library(scater)
```

```r
umi <- readRDS("completely_normalised_data.scran.combat_glm.sc3_clust.rds")
```

```r
message("Let's see how these k clustering fit with our PCAs. (Note that scater stores only the
first two PCs)")

umi <- runPCA(umi, ntop = 10000, ncomponents = 18)

pdf(file = "Rplots%05d.pdf", 60,60, FALSE)
for (k in 2:10){
    print(plotPCA(umi, ncomponents = 18, colour_by = sprintf("sc3_%d_clusters", k)))
}
dev.off()

```
